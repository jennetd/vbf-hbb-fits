```python
import ROOT as ROOT
import numpy as np
```

```python
year = "2018"
```

```python
line = ROOT.TGraph(2)
line.SetPoint(0,40,0)
line.SetPoint(1,201,0)
line.SetLineColor(2)
line.SetLineWidth(4)
```

```python
p = "vbfhi"
f = ROOT.TFile("output/testModel_qcdfit_"+p+"_"+year+".root")

w = f.Get("w")
mc = w.obj("ModelConfig")

data_pass1 = w.data("ptbin0"+p+"pass"+year+"_data_obs")
data_fail1 = w.data("ptbin0"+p+"fail"+year+"_data_obs")

c1 = ROOT.TCanvas("c_"+p,"c_"+p,600,600)
frame1 = w.var("msd").frame(23)
ROOT.gPad.SetLeftMargin(0.15)

fit1 = w.pdf("ptbin0"+p+"pass"+year+"_qcd")
fitplot1 = fit1.plotOn(frame1,ROOT.RooFit.LineColor(2))
cfit = fitplot1.getCurve()

hpass1 = data_pass1.plotOn(frame1, ROOT.RooFit.Rescale(1.0/data_pass1.sumEntries()), 
                          ROOT.RooFit.DataError(1),ROOT.RooFit.LineColor(1),ROOT.RooFit.MarkerColor(1))
hpass = hpass1.getHist().makeResidHist(cfit)
hfail1 = data_fail1.plotOn(frame1, ROOT.RooFit.Rescale(1.0/data_fail1.sumEntries()), 
                          ROOT.RooFit.DataError(1),ROOT.RooFit.LineColor(4),ROOT.RooFit.MarkerColor(4))
hfail = hfail1.getHist().makeResidHist(cfit)

c1.Draw()
frame1.SetMaximum(0.3)
frame1.SetMinimum(0)
frame1.SetTitle("VBF high m_{jj} bin, "+year)
frame1.SetYTitle("Events / 7 GeV")
frame1.SetXTitle("m_{sd} [GeV]")
frame1.Draw()
```

```python
c = ROOT.TCanvas("c_"+p,"c_"+p,600,600) 
ROOT.gPad.SetLeftMargin(0.15)

hpass.SetTitle("VBF high m_{jj} bin, "+year)
hpass.GetYaxis().SetTitle("Residuals")
hpass.GetXaxis().SetTitle("m_{sd} [GeV]")
hpass.Draw("ap")
hfail.SetLineColor(4)
hfail.SetMarkerColor(4)
hfail.Draw("psame")
line.Draw("csame")

c.Draw()
```

```python
p = "ggf"
f = ROOT.TFile("output/testModel_qcdfit_"+p+"_"+year+".root")

w = f.Get("w")
mc = w.obj("ModelConfig")

canvases = []
frames = []

failhists = []
passhists = []

for i in range(0,5):

    data_pass = w.data("ptbin"+str(i)+p+"pass"+year+"_data_obs")
    data_fail = w.data("ptbin"+str(i)+p+"fail"+year+"_data_obs")

    c = ROOT.TCanvas("c_"+p+str(i),"c_"+p+str(i),600,600)
    frame = w.var("msd").frame(23)
    ROOT.gPad.SetLeftMargin(0.15)
    
    fit = w.pdf("ptbin"+str(i)+p+"pass"+year+"_qcd")
    fitplot = fit.plotOn(frame,ROOT.RooFit.LineColor(2))
    cfit = fitplot.getCurve()

    hpass0 = data_pass.plotOn(frame, ROOT.RooFit.Rescale(1.0/data_pass.sumEntries()), 
                              ROOT.RooFit.DataError(1),ROOT.RooFit.LineColor(1),ROOT.RooFit.MarkerColor(1))
    hpass = hpass0.getHist().makeResidHist(cfit)
    hfail0 = data_fail.plotOn(frame, ROOT.RooFit.Rescale(1.0/data_fail.sumEntries()), 
                              ROOT.RooFit.DataError(1),ROOT.RooFit.LineColor(4),ROOT.RooFit.MarkerColor(4))
    hfail = hfail0.getHist().makeResidHist(cfit)
    
    failhists += [hfail]
    passhists += [hpass]
    
    canvases += [c]
    frames += [frame]
    
    canvases[i].Draw()
    frames[i].SetMaximum(0.2)
    frames[i].SetMinimum(0)
    frames[i].SetTitle("ggF p_{T} bin "+str(i+1)+", "+year)
    frames[i].SetYTitle("Events / 7 GeV")
    frames[i].SetXTitle("m_{sd} [GeV]")
    frames[i].Draw()
    
```

```python
canvases = []

for i in range(0,5):
    
    c = ROOT.TCanvas("c_"+p+str(i),"c_"+p+str(i),600,600)
    ROOT.gPad.SetLeftMargin(0.15)
    
    hpass = passhists[i]
    hfail = failhists[i]
    
    hpass.SetTitle("ggF p_{T} bin "+str(i+1)+", "+year)
    hpass.GetYaxis().SetTitle("Residuals")
    hpass.GetXaxis().SetTitle("m_{sd} [GeV]")
    hpass.Draw("ap")
    hfail.SetLineColor(4)
    hfail.SetMarkerColor(4)
    hfail.Draw("psame")
    line.Draw("csame")
    
    canvases += [c]
    canvases[i].Draw()
```
